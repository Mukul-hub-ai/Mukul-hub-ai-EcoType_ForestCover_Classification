# outlier detection and handling 

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# find the numerical columns for outlier
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
print("All numerical columns:")
print(numerical_cols)
```

```python
# identify binary and numerical columns
binary_cols = []
continuous_numerical_cols = []

for col in numerical_cols:
    unique_vals = df[col].nunique()
    unique_values_list = df[col].unique()
    
    # Check if binary column (only 0 and 1 values)
    if unique_vals == 2 and set(unique_values_list).issubset({0, 1}):
        binary_cols.append(col)
    else:
        continuous_numerical_cols.append(col)

print("Continuous Numerical Columns (Outlier treatment needed):\n")
print(continuous_numerical_cols)
print(f"\nTotal continuous numerical columns: {len(continuous_numerical_cols)}")

print("\nBinary Columns (NO outlier treatment needed):\n")
print(binary_cols)
print(f"\nTotal binary columns: {len(binary_cols)}")
```

```python
numerical_cols = [
    'Elevation', 'Aspect', 'Slope',
    'Horizontal_Distance_To_Hydrology', 'Vertical_Distance_To_Hydrology',
    'Horizontal_Distance_To_Roadways', 'Hillshade_9am', 'Hillshade_Noon',
    'Hillshade_3pm', 'Horizontal_Distance_To_Fire_Points'
]  
plt.figure(figsize=(20, 10))
plt.suptitle("Boxplots Before Outlier Treatment - Forest Cover Dataset", fontsize=16, fontweight='bold')
for i, col in enumerate(numerical_cols, 1):
    plt.subplot(2, 5, i)     # 2 rows and 5 columns 
    sns.boxplot(y=df[col], color='skyblue')
    plt.title(f'{col}', fontweight='bold')
    plt.tight_layout()

plt.show()
```

```python
import numpy as np

# create a copy of original dataset
df_clean = df.copy()

for col in numerical_cols:
    Q1 = df_clean[col].quantile(0.25)  # 25th percentile
    Q3 = df_clean[col].quantile(0.75)  # 75th percentile
    IQR = Q3 - Q1  # interquartile range

    # Define lower and upper bound
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Count outliers before treatment
    outliers_before = ((df_clean[col] < lower_bound) | (df_clean[col] > upper_bound)).sum()

    print(f"{col}: {outliers_before} outliers before")
    
     # Outlier capping using clip
    df_clean[col] = df_clean[col].clip(lower_bound, upper_bound)
    
print("\nOutlier treatment completed")
```

```python
numerical_cols = [
    'Elevation', 'Aspect', 'Slope',
    'Horizontal_Distance_To_Hydrology', 'Vertical_Distance_To_Hydrology',
    'Horizontal_Distance_To_Roadways', 'Hillshade_9am', 'Hillshade_Noon',
    'Hillshade_3pm', 'Horizontal_Distance_To_Fire_Points'
]  
plt.figure(figsize=(20, 10))
plt.suptitle("Boxplots After Outlier Treatment - Forest Cover Dataset", fontsize=16, fontweight='bold')
for i, col in enumerate(numerical_cols, 1):
    plt.subplot(2, 5, i)     # 2 rows and 5 columns 
    sns.boxplot(y=df_clean[col], color='skyblue')
    plt.title(f'{col}', fontweight='bold')
    plt.tight_layout()

plt.show()
```

# Skewness Check & Transformation 

```python
for col in numerical_cols:
    skew = df_clean[col].skew()
    print(f"Skewness for {col}: {skew}")

    if abs(skew) > 1:
        df_clean[col] = np.log1p(df_clean[col])
```

```python
# Reset index
df_clean = df_clean.reset_index(drop = True)
# Final Summary
print("\nData cleaning done. Shape after cleaning:", df_clean.shape)
print("First few rows after cleaning:\n", df_clean.head())
```

# Feature Engineering 

```python
# 1 Hydrology Distance Ratio
df_clean['Hydrology_Distance_Ratio'] = df_clean['Vertical_Distance_To_Hydrology'] / (df_clean['Horizontal_Distance_To_Hydrology'] + 1)

# 2 Road to Hydrology Ratio  
df_clean['Road_to_Hydrology_Ratio'] = df_clean['Horizontal_Distance_To_Roadways'] / (df_clean['Horizontal_Distance_To_Hydrology'] + 1)

# 3 Fire to Hydrology Ratio
df_clean['Fire_to_Hydrology_Ratio'] = df_clean['Horizontal_Distance_To_Fire_Points'] / (df_clean['Horizontal_Distance_To_Hydrology'] + 1)

# 4 Hillshade Morning vs Noon Difference
df_clean['Hillshade_Morning_Noon_Diff'] = df_clean['Hillshade_9am'] - df_clean['Hillshade_Noon']

# 5 Hillshade Noon vs Evening Difference  
df_clean['Hillshade_Noon_Evening_Diff'] = df_clean['Hillshade_Noon'] - df_clean['Hillshade_3pm']

# 6 Hillshade Daily Variation (Absolute difference)
df_clean['Hillshade_Daily_Variation'] = abs(df_clean['Hillshade_9am'] - df_clean['Hillshade_3pm'])
```

# Model Building 

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix, classification_report
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from xgboost import XGBClassifier
```

# Load data

```python
df = pd.read_csv("forest_cover_top20_features.csv")
X = df.drop('Cover_Type', axis=1)
y = df['Cover_Type']
df
```

```python
# Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
```

# Models

```python
models = {
    "Random Forest": RandomForestClassifier(random_state=42),
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "Logistic Regression": LogisticRegression(max_iter=1000, random_state=42),
    "KNN": KNeighborsClassifier(n_neighbors=5),
    "XGBoost": XGBClassifier(random_state=42)
}
```

```python
# Train and Evaluate Models
results = {}

print("TRAINING MODELS...")
for name, model in models.items():
    print(f"\nTraining {name}...")
    
    # Train model
    model.fit(X_train, y_train)
    
    # Make predictions
    y_pred = model.predict(X_test)
    
    # Calculate metrics
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, average='weighted', zero_division=0)
    recall = recall_score(y_test, y_pred, average='weighted', zero_division=0)
    cm = confusion_matrix(y_test, y_pred)  
    cr = classification_report(y_test, y_pred)
    
    # Store results
    results[name] = {
        'Accuracy': accuracy,
        'Precision': precision,
        'Recall': recall,
        'Confusion_Matrix': cm,  
        'Classification_Report': cr
    }
    
    # Print results
    print(f" {name}")
    print(f"   Accuracy:  {accuracy:.4f}")
    print(f"   Precision: {precision:.4f}")
    print(f"   Recall:    {recall:.4f}")
    print(f"   Confusion Matrix:") 
    print(cm)  
    print(f"   Classification Report:")  
    print(cr) 
    print("-" * 50) 

```

```python
# Create Comparison DataFrame
results_df = pd.DataFrame(results).T
print("\n" + "="*60)
print("MODEL COMPARISON")
print("="*60)
print(results_df[['Accuracy', 'Precision', 'Recall']].round(4))

# Best model
best_model = results_df['Accuracy'].idxmax()
print(f"\n BEST MODEL: {best_model}")
print(f" Accuracy:  {results_df.loc[best_model, 'Accuracy']:.4f}")
print(f" Precision: {results_df.loc[best_model, 'Precision']:.4f}")
print(f" Recall:    {results_df.loc[best_model, 'Recall']:.4f}")

# Display detailed results for best model
print(f"\n DETAILED RESULTS FOR BEST MODEL ({best_model}):")
print("Confusion Matrix:")
print(results[best_model]['Confusion_Matrix'])
print("\nClassification Report:")
print(results[best_model]['Classification_Report'])
```

# Hyperparameter Tuning

```python
# KNN Hyperparameter Tuning using RandomizedSearchCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix, classification_report
import numpy as np
import pandas as pd

print("KNN HYPERPARAMETER TUNING WITH RANDOMIZEDSEARCHCV")

# Define KNN model
knn = KNeighborsClassifier()

# Define parameter distributions for KNN
param_dist = {
    'n_neighbors': list(range(1, 31)),           # 1 to 30 neighbors
    'weights': ['uniform', 'distance'],          # Weighting strategy
    'algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute'],  # Algorithm type
    'leaf_size': [10, 20, 30, 40, 50],          # Leaf size for tree algorithms
    'p': [1, 2]                                 # 1=Manhattan, 2=Euclidean distance
}

# Perform RandomizedSearchCV
random_search = RandomizedSearchCV(
    estimator=knn,
    param_distributions=param_dist,
    n_iter=50,                    # Number of parameter combinations to try
    cv=5,                         # 5-fold cross-validation
    scoring='accuracy',           # Optimize for accuracy
    random_state=42,              # Reproducible results
    n_jobs=-1,                    # Use all processors
    verbose=1                     # Show progress
)

# Fit the model
print("Training KNN with RandomizedSearchCV...")
random_search.fit(X_train, y_train)
```

```python
# Get best results
best_params = random_search.best_params_
best_score = random_search.best_score_
best_knn_model = random_search.best_estimator_

print(f"\n BEST PARAMETERS FOUND:")
for param, value in best_params.items():
    print(f"  {param}: {value}")

print(f"BEST CROSS-VALIDATION ACCURACY: {best_score:.4f}")
```

```python
# Evaluate tuned model on test set
y_pred_tuned = best_knn_model.predict(X_test)
tuned_accuracy = accuracy_score(y_test, y_pred_tuned)
tuned_precision = precision_score(y_test, y_pred_tuned, average='weighted', zero_division=0)
tuned_recall = recall_score(y_test, y_pred_tuned, average='weighted', zero_division=0)

print(f"\nTUNED KNN PERFORMANCE ON TEST SET:")
print(f"Accuracy:  {tuned_accuracy:.4f}")
print(f"Precision: {tuned_precision:.4f}")
print(f"Recall:    {tuned_recall:.4f}")
```

```python
# Compare with original KNN model
try:
    original_accuracy = results['KNN']['Accuracy']
    improvement = tuned_accuracy - original_accuracy
    print(f"\nCOMPARISON WITH ORIGINAL KNN:")
    print(f"Original KNN Accuracy: {original_accuracy:.4f}")
    print(f"Tuned KNN Accuracy:    {tuned_accuracy:.4f}")
    print(f"Improvement:           {improvement:.4f}")

    if improvement > 0:
        print("Hyperparameter tuning improved KNN performance!")
    else:
        print("KNN performance maintained with optimized parameters.")
except:
    print(f"\nTuned KNN Test Accuracy: {tuned_accuracy:.4f}")
```

```python
# Display confusion matrix and classification report
print(f"\nDETAILED RESULTS:")
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_tuned))
print("\nClassification Report:")
print(classification_report(y_test, y_pred_tuned))
```
